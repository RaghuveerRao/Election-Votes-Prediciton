#### Building Predictive Model
Random Forest, Decision Tree and SVM were tried, settled on Random Forest Regression for best accuracy

Grid Search for hyperparameter tuning

```python
parameters = {'max_depth':range(2,9), 'min_samples_split':range(2,6),'min_samples_leaf':range(1,6) , 'n_estimators': [50,200,500]}
clf = GridSearchCV(RandomForestRegressor(), parameters, n_jobs=-1, cv=3, scoring='r2')
clf.fit(X=x, y=y)
```

```python
tree_model = clf.best_estimator_
print (clf.best_score_, clf.best_params_) 
```

Using tuned hyper parameters for prediction

```python
regr = RandomForestRegressor(max_depth=8 ,min_samples_leaf= 1, min_samples_split= 5, random_state=0, n_estimators=50)
fit = regr.fit(x,y)
y_pred = fit.predict(x)
r2_score(y,y_pred)
```

```python
mean_absolute_error(y,fit.predict(x))
```

Extracting feature importance

```python
imp = fit.feature_importances_
idx = (-imp).argsort()[:20]
for i in range(0,20):
    print(x.columns[imp == imp[idx][i]])
```
